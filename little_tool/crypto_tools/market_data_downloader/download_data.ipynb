```python
import ccxt
import datetime
import pandas as pd

binance = ccxt.binance()
```

```python
symbol = 'BTC/USDT' # 幣別
timeframe = '1h' # 時間區間
start_time = '2018-01-01 00:00:00' # 開始時間
end_time = '2018-01-03 23:59:59' # 結束時間

start_time = datetime.datetime.strptime(start_time, '%Y-%m-%d %H:%M:%S')
end_time = datetime.datetime.strptime(end_time, '%Y-%m-%d %H:%M:%S')

start_time = int(start_time.timestamp() * 1000)
end_time = int(end_time.timestamp() * 1000)

for t in range(start_time, end_time, 2000 * 60 * 60):
    ohlcv = binance.fetch_ohlcv(symbol, timeframe, t, 2000)
    df = pd.DataFrame(ohlcv, columns=['time', 'open', 'high', 'low', 'close', 'volume'])
    df['time'] = pd.to_datetime(df['time'], unit='ms')
    df.to_csv('data.csv', mode='a', header=False, index=False)

df = pd.read_csv('data.csv', names=['time', 'open', 'high', 'low', 'close', 'volume'])

df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')
df = df.sort_index()

df
```
